## 1. 언어
1) 오타 교정/검색 엔진 서비스
  * Grammarly의 문장 교정 서비스
    - 사용자가 오타가 포함된 어휘나 문법에 어긋난 문장을 입력하면 자동으로 교정하는 서비스
    - grammarly 서비스는 규칙, 패턴, 기계학습, 딥러닝, 자연어 처리 방식 등의 AI 방법론을 조합하여 구현한 프로그램으로 글쓰기 능력을 향상시켜 줌
    - grammarly AI시스템은 피드백 매커니즘도 갖추고 있어서, 특정 제안을 많은 사용자가 무시하면 더 나은 제안으로 바꿀 수 있도록 알고리즘을 조정한다.


2) 챗봇
  * 와이즈넛 
    - 교육 행정용 대화형 로봇 시스템 : 대학 챗봇은 입시, 장학, 수업, 취업 등 대학의 행정 서비스 관련 문의를 24시간 365일 자동으로 응대하는 대화형 서비스
    - '휴학을 신청하고 싶어요'라고 질문하면 휴학 신청 방법, 신청 절차, 휴학 시 등록금 처리 방법 등 모든 관련 정보를 제공
  * 카카오엔터프라이즈 & 교보생명 챗봇 AI 기술 도입

3) 번역
  * 구글 번역기
  * 네이버 파파고

4) HR 영역
  * 서류 검토 및 면접에 AI 기술을 도입하여 인적 자원 관리가 실시간으로 이뤄지도록 함. 수시로 뽑고, 수시로 평가하고, 수시로 이동하는 구조가 만들어 질 것으로 예상됨

## 3. 이미지
1) 패션
  * 종달랩 
    - 패션 의류 원단, 부자재 검색 
    - AI 기반 이미지 검색 엔진으로 명칭이 없어 찾기 힘든 원단, 부자재를 사진으로 검색
  * 옴니어스
    - 이미지 인식, 고객 취향에 맞는 패션 상품 추천
  * 아마존의 스타일 스냅
    - 의류 검색 서비스
    - 고객이 자신의 사진이나 원하는 이미지 등을 업로드하면 인공지능이 사진에 있는 모습과 비슷하거나 취향이 비슷한 의류를 제시해 줌
    - 컴퓨터 비전과 딥러닝을 사용하여 사진의 의류 항목을 식별하고, 딥러닝으로 이미지의 의류 아이템을 '착용감이 조흔 드레스' 또는 '플란넬 셔츠'와 같은 범주로 분류
  * 삼성물산 AI 아이피츠
    - SNS 분석을 통해 유행할 옷이나 색을 예측한 디자인 초안을 제시, 재고 물량 조정 업무

2) 헬스 케어
  * 진료
    - 딥러닝을 통해 대량의 의료 영상기록을 처리함으로써 의료진의 치료 결정에서의 불확실성 감소
    - 대규모 의료 영상을 빠르게 처리해 질환 형태, 음성/양성 판단 등에 적용
    - 진료 기록과 같은 긴 서술형 문자 묶음들을 인식하여 행정 기록에 남길 수 있게 변환
    - 환자들의 대용량 의료 데이터를 빠르게 분석하여 환자의 치료 결과 예측
    - 헬스케어 기관들이 보유한 방대한 환자 의료데이터를 통해 환자와 치료 제공자들에게 맞춤형 권고를 제공
  * 로보틱스
    - 수술 과정의 정밀함과 정확도를 높여 질 높은 치료를 제공

 ![SAE 자율 주행 단계](http://img.newspim.com/news/2019/10/15/1910151547134440_w.jpg)

  * 테슬라 오토파일럿
    - 현재 모든 테슬라 신차에는 오토파일럿 기능을 위한 첨단 하드웨어가 기본으로 탐재되어 있으며, 미래에는 소프트웨어 업데이트와 기능 개선을 통하여 완전 자율 주행 능력을 갖추게 될 것
    - 현재 오토파일럿 기능은 운전자의 제어가 필요하며 완전 자율 주행이 가능한 것은 아님
    - 완전 자율 주행 기능 목표
      + 차에 타서 목적지를 말하기만 하면 됩니다. 아무 말도 하지 않으면 자동차는
달력을 확인해서 가정한 목적지로 가거나, 달력에 일정이 따로 없으면 집으로 갑니다.
Tesla는 최적의 경로를 계산해서 도시 거리 (차선 표시가 없어도)를 주행하며,
신호등, 정지 신호 및 로터리가 있는 복잡한 교차로도 주행할 수 있고, 많은 차량들이 빠르게 이동하는 밀집되고 혼잡한 고속도로에서도 주행할 수 있습니다. 목적지에 도착해서 입구에 내리면 차량이 주차 탐색 모드로 진입한 후,
자동으로 주차 공간을 찾아서 주차합니다 핸드폰을 탭해서 다시 차량을 부를 수 있습니다.
