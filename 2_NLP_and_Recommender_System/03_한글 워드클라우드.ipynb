```python
ktext = ktext.replace('\n', '')
```

```python
# Tokenize
words = Okt().morphs(ktext)
```

```python
# 불용어 제거
stopwords = ['에', '게', '의', '은', '는', '이', '가', '이다', '인', '것',
             '을', '를', '수', '.', '\n', ',', '그리고', '에서', '들', '한다',
             '같은', '도', '와', '과', '로', '거', '건', '(', ')', '그', '하는', '이라는']

words_sw_removed = words.copy()

for word in words:
    if word in stopwords:
        words_sw_removed.remove(word)
        

print('전체 단어 모음(', str(len(words)), '):', words)
print('-' * 80)
print('불용어 제거한 단어 모음(', str(len(words_sw_removed)), '):', words_sw_removed)
```

```python
# Counter
count_list = Counter(words_sw_removed)
```

```python
my_wc = WordCloud(font_path='C:/Users/gjai_kmj/AppData/Local/Microsoft/windows/Fonts/SDSwaggerTTF.ttf', background_color='white')
plt.figure(figsize=(12, 8))
plt.imshow(my_wc.generate_from_frequencies(count_list))
plt.axis('off')
print()
```
