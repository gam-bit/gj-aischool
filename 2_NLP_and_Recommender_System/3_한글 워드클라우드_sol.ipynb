```python
#   
from konlpy.tag import Okt  
#     
from collections import Counter
# 
from wordcloud import WordCloud
import matplotlib.pyplot as plt


ktext = '''
      
 ` `       .
       ` `   ,
    ,      .
        ,  ,  
 ~     .
   ?
        
     ,
     .
   (?) .;;
         .
mj        .
 `   `...   .
    
       .
      .
        .
   
,        .
'''

words = Okt().morphs(ktext)

sw = ['.', ',', "'", '\n', '', '', '', '']

sw_removed = []
for i in words:
    if i.lower() not in sw:
        sw_removed.append(i)

count_list = Counter(sw_removed)

my_wc = WordCloud(font_path='C:/Users/gjai_kmj/AppData/Local/Microsoft/windows/Fonts/SDSwaggerTTF.ttf', background_color='white')
plt.imshow(my_wc.generate_from_frequencies(count_list))
plt.axis('off')
plt.show()
```
