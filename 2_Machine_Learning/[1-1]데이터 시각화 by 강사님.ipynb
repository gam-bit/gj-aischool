#### 그림 범위 지정

- 플롯 그림을 보면 몇몇 점들은 그림의 범위 경계선에 있어서 잘 보이지 않는 경우
- 그림의 범위를 수동으로 지정하려면 xlim 명령과 ylim 명령을 사용
- 그림의 범위가 되는 x축, y축의 최소값과 최대값을 지정


````python
plt.xlim(-2,10)
plt.ylim(-5,12)
plt.plot(x, y, ls='--', marker='o', ms=5, mec='g', mew=1, mfc='g')
````

#### 틱(tick)

````python
plt.xticks([0,3,6])
plt.yticks([1,5,9])

plt.plot(x, y, ls='--', lw=5, marker='o', ms=15, mec='g', mew=5, mfc='r')
````

````python
# label 이 잘못 될 때를 비교 

# 카테고리 값 : 
bins = [-1,2,5,8,11,14,17,20,24]
labels = ['0-2', '3-5', '6-8', '9-11', '12-14', '15-17', '18-20', '21-23']

time_df = ( data['발생년월일시'] % 100 )
rs = pd.cut(time_df.values, bins,labels=labels).value_counts()

t = [x % 100 for x in data['발생년월일시']]
time_df2 = pd.DataFrame(t, columns = ['사고시간'])
rs2 = pd.cut(time_df2['사고시간'], bins,labels=labels).value_counts()
 
rs3 = pd.cut(time_df2['사고시간'].values, bins,labels=labels).value_counts()


# rs, rs2, rs3 파이차트로 비교
plt.figure(figsize =(5,5))

# 첫번째 2는 열의 개수를 가리킨다. 두번째 2는 행의 개수, 
# 그리고 마지막 1은 4개의 자리 중 첫번째 자리란 뜻
# plt.subplot(2, 2, 1)이라 써도 결과는 같다.
plt.subplot(221)
plt.pie(rs, labels=labels, autopct='%1.1f%%')

plt.subplot(222)
plt.pie(rs2, labels=labels, autopct='%1.1f%%')

plt.subplot(223)
plt.pie(rs3, labels=labels, autopct='%1.1f%%')
plt.show()
````

````python
# 파이차트
labels = ('A', 'B', 'C') 
xs = [50, 30, 40] 

patches, texts, autotexts = plt.pie(
    labels=labels, 
    labeldistance=1.1,   # label과 파이의 간격, 1일 경우 딱 붙음. 
    x = xs,              # 값
    explode=(0.1, 0.1, 0), # 파이조각 돌출 크기 (0 이면 돌출되지 않음)
    startangle=90,       # 그려지는 파이 조각 시작위치. 90은 12시 방향 
    shadow=False,        # 파이차트 그림자 효과 유무
    counterclock=False,  # 파이차트가 그려질 방향 (시계방향, 반시계 방향) 
    autopct='%1.1f%%',   # 파이조각 위에 표시될 글자 형태 (파이조각 전체대비 백분율)
    pctdistance=0.7,     # 백분률 표시 위치 (radius 기준으로 위치 선정)
    colors=['grey', 'red', 'blue'],  # 각 파이조각 컬러
)
````
