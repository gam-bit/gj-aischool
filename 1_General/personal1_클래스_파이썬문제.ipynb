2. 이름과 동물을 지정해서 인스턴스를 생성하는 동물이라는 클래스를 생성하고, 동물을 상속받아 고양이 클래스를 만들어 야옹을 출력하는 함수를 만들고, 동물을 상속받아 강아지 클래스를 만들어 왈왈을 출력하는 함수를 만들어주세요.

```
# 1

class Rectangle:
  def __init__(self, width, height):
    self.width = width
    self.height = height
    print('직사각형이 생성되었습니다.')
    
  def area(self):
    return self.width * self.height

  def circum(self):
    return 2 * (self.width + self.height)


#-----------------

rec1 = Rectangle(5, 4)
print(f'넓이: {rec1.area()}')
print(f'둘레: {rec1.circum()}')
```

```
# 2

class Pet:
  
  def __init__(self):
    print('펫 생성!')


class Cat(Pet):
  
  def bark(self):
    print('야옹')


class Dog(Pet):

  def bark(self):
    print('왈왈')


#------------------------

cat1 = Cat()
cat1.bark()

dog1 = Dog()
dog1.bark()
```

# 파이썬 문제(출처: 백준 알고리즘)

1. 피보나치 수열의 코드는 다음과 같습니다.(global, for, list와 append, input, range, def 사용)


```
def fibonacci(n):
  if n == 0:
    return 0
  elif n == 1:
    return 1
  else:
    return fibonacci(n-1) + fibonacci(n-2)
```

실제 수열은 1 1 2 3 5 8 13 ... 으로 진행됩니다.


수열의 1번째 수는 1이 return되고, 수열의 2번째 수는 0 한번, 1 한번, 3번째 수는 0 한번, 1 두번 return 됩니다.

수열의 n번째 수가 0, 1을 총 몇 번 리턴하는지 구하세요.

```
def exe(n): # n > 1
  
  s = [[0], [1]]

  for i in range(2, n+1):
    s.append(s[i-1] + s[i-2])

  return s[n].count(0), s[n].count(1)

#----------------------------------

exe(3)
     
```

2. 별 찍기(for, range, input사용)

5를 입력하면, 아래와 같이 출력이 됩니다. 규칙을 유추해서 다음과 같은 규칙으로 출력하는 함수를 구하세요.

```
*********
 *******
  *****
   ***
    *
   ***
  *****
 *******
*********
```

```
def star(n):
  
  stars = "{:^" + str(2*n-1) + "}"

  for i in range(1,n+1):    
    print(stars.format('*'*(2*n-2*i+1)))
  
  for i in range(1, n):
    print(stars.format('*'*(2*i+1)))

#-------------------------------

star(5)
```
